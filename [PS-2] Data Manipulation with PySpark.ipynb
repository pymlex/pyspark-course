## Schemas

A `schema` is the blueprint of your data. It defines the names of your columns and their data types, like whether a value should be a string, integer, or timestamp. In Spark, having a defined schema is a huge plus because it prevents errors and saves the system from having to guess (infer) the types, which makes processing faster and more reliable.



```python
df.printSchema()
```

Schema creation with `SQL`-like syntax.

```python
my_ddl_schema = '''
    Item_Identifier string,
    Item_Weight STRING,
    Item_Fat_Content string,
    Item_Visibility double,
    Item_Type string,
    Item_MRP double,
    Outlet_Identifier string,
    Outlet_Establishment_Year int,
    Outlet_Size string,
    Outlet_Location_Type string,
    Outlet_Type string,
    Item_Outlet_Sales double
'''
df = spark.read.format('csv') \
    .schema(my_ddl_schema) \
    .option('header', True) \
    .load('/Volumes/workspace/default/study/BigMart Sales.csv')
df.printSchema()
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

`Python`-style schema.

```python
my_struct = StructType([
    StructField('Item_Identifier', StringType(), True),
    StructField('Item_Weight', StringType(), True),
    StructField('Item_Fat_Content', StringType(), True),
    StructField('Item_Visibility', DoubleType(), True),
    StructField('Item_Type', StringType(), True),
    StructField('Item_MRP', DoubleType(), True),
    StructField('Outlet_Identifier', StringType(), True),
    StructField('Outlet_Establishment_Year', IntegerType(), True),
    StructField('Outlet_Size', StringType(), True),
    StructField('Outlet_Location_Type', StringType(), True),
    StructField('Outlet_Type', StringType(), True),
    StructField('Item_Outlet_Sales', DoubleType(), True)
])
df = spark.read.format('csv') \
    .schema(my_struct) \
    .option('header', True) \
    .load('/Volumes/workspace/default/study/BigMart Sales.csv')
df.printSchema()
```

## Writing

Write `CSV`. You can apply different modes:

 - `ignore` — do nothing if exists.
 - `error` — abort if exists.
 - `overwrite` — replace.
 - `append` — add rows.

```python
df.write.format('csv') \
    .mode('overwrite') \
    .save('/Volumes/workspace/default/study/saved.csv')
```

Save in `Delta` format. Writing `CSV` may lead to errors if the file is being processed by another program. `Delta` automatically saves the schema, each operation is either completed and logged or `rolled back` if an error occurs. It also supports `versioning`, while `CSV` simply overwrites files, which can lead to accidental data loss. Furthermore, `Delta` stores data statistics, allowing `Spark` to skip files that do not contain the required rows, significantly improving performance. 

That is why `Delta` satisfies `ACID` properties: 

- `Atomicity` — the operation succeeds completely or fails entirely.
- `Consistency` — ensures data matches the schema and rules.
- `Isolation` — multiple processes won't interfere with each other.
- `Durability` — changes are permanent after the commit, even if the system crashes.

```python
df.write.format('delta') \
    .mode('overwrite') \
    .saveAsTable('my_table')
```

## Basic operations

Selection.

```python
df_sel = df.select(
    col('Item_Identifier').alias('ITEM_ID'), 
    col('Item_Weight'), 
    col('Item_Fat_Content')
)
df_sel.limit(10).display()
```

Filter with a complex condition.

```python
df_sel = df.filter((col('Outlet_Size').isNull()) & \
                   (col('Item_Fat_Content') == 'Regular') & \
                   (col('Item_Weight') < 10) & \
                   (col('Outlet_Location_Type').isin('Tier 1', 'Tier 2'))
                   )
df_sel.limit(10).display()
```

Exactly rename a column, not just make an alias

```python
df_sel = df.withColumnRenamed('Item_Weight', 'WEIGHT')
df_sel.limit(10).display()
```

Create a column where each cell contains a fixed string value.

```python
df = df.withColumn('flag', lit('some_flag'))
```

Multiplication of two columns.

```python
df = df.withColumn('multiply', col('Item_Weight') * col('Item_MRP'))
df.limit(10).display()
```

Explain lazy operations even before they are performed.

```python
df_a = df.select(col('Item_Weight') + 10 * col('Item_Visibility')) \
  .filter(col('Item_Type') == 'Meat')
df_a.explain()
df_a.limit(10).display()
```

## String operations

Substring replacement.

```python
df = df.withColumn('Item_Fat_Content', regexp_replace(col('Item_Fat_Content'), 'Regular', 'Reg'))
df.limit(10).display()
```

Capitalisation.

```python
df.select(initcap(col('Item_Type')).alias('str_oper')).limit(10).display() #upper, lower
```

Data types conversion.

```python
df = df.withColumn('Item_weight', col('Item_weight').cast(StringType()))
df.printSchema()
```

## Union

Concatenate Dataframes by column name.

```python
data1 = [
    ('dfsdf', '1'),
    ('fdsfsd', '2')
]
schema1 = """
    name STRING,
    id STRING
"""
schema2 = """
    id STRING,
    name STRING
"""
data2 = [
    ('3', 'eef'),
    ('4', 'gtfgfdsd')
]

df1 = spark.createDataFrame(data1, schema1)
df2 = spark.createDataFrame(data2, schema2)
```

```python
df1.unionByName(df2).display()
```

## Process dates

Get the current date.

```python
df = df.withColumn('curr_date', current_date())
df.limit(10).display()
```

Add days to a date.

```python
df = df.withColumn('week_after', date_add(col('curr_date'), 7)) #date_sub
df.limit(10).display() 
```

Difference between two dates.

```python
df = df.withColumn('datediff', datediff(col('week_after'), col('curr_date')))
df.limit(10).display()
```

Date formatting.

```python
df = df.withColumn('week_after', date_format('week_after', 'dd.mm.yy'))
df.limit(10).display()
```

## User defined functions (UDF)

You can define your own functions but make sure `Spark` needs to call `Python` code which slows down the process.

```python
def my_f(x):
  return x ** 2

my_udf = udf(my_f)
df.withColumn('square', my_udf(col('Item_MRP'))).limit(10).display()
```

## Several values processing

Split by space.

```python
df.withColumn('Outlet_Type', split('Outlet_Type', ' ')[0]).limit(10).display()
```

Split into several rows.

```python
df_explode = df.withColumn('Outlet_Type', split('Outlet_Type', ' '))
df_explode.withColumn('Outlet_Type', explode('Outlet_Type')).limit(10).display()
```

Check if a cell with an array contains a value.

```python
df_explode.withColumn('Type1_flag', array_contains('Outlet_type', 'Type1')).limit(10).display()
```

Aggregation.

```python
df.groupBy('Item_Type', 'Outlet_Size').agg(
  sum('Item_MRP').alias('Total'), 
  avg('Item_MRP').alias('Avg')
).limit(10).display()
```

## 1NF violation

First Normal Form (`1NF`) ensures each table cell contains only one atomic value, meaning no lists or multiple entries in a single field. Every row must be unique, and the order of data should not affect its meaning.

```python
data3 = [
    ('user1', 'fdsfa'),
    ('user1', 'edfsd'),
    ('user2', 'pdsf')
]
schema3 = """
    user STRING, 
    book STRING
"""
df3 = spark.createDataFrame(data3, schema3)
df3.display()
```

```python
df3.groupby('user').agg(collect_list('book').alias('books')).display()
```
